```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

#load data

```python
data_train = "https://docs.google.com/spreadsheets/d/e/2PACX-1vSG8gRBml0Y1N8nxlFGkUyG4bp0gFllEobzC7nd1fM6TjHdWFHTn_GPEKcQWb_SEcuDaEDflX_14hZ2/pub?gid=1270430478&single=true&output=csv"

train = pd.read_csv(data_train)
train
```

#data understanding

```python
train.head()
```

```python
train.info()
```

```python
train.isnull().sum()
```

```python
train1 = pd.DataFrame(train.isnull().sum().sort_values(ascending=False))
print(train1.to_markdown())
```

```python
# drop column
train = train.drop(['DC206','DC207','DC208','DC209','DC213','DC214','DC215','DC216','DC220','DC230a','DC230b','DC232','DC232b','DC235','DC241','DC242','DC244','DC246','DC252','DC142a'],axis=1)
train.isna().sum()
```

```python
# drop row
train = train.dropna(subset = ['DC201','DC205'])
train.isna().sum()
```

```python
# fill with mode
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='most_frequent', missing_values=np.nan)
train[['DC210','DC211','DC212','DC226','DC237','DC237a','DC237b','DC237c','DC237d','DC237e','DC237f','DC109','DC226']] = imputer.fit_transform(train[['DC210','DC211','DC212','DC226','DC237','DC237a','DC237b','DC237c','DC237d','DC237e','DC237f','DC109','DC226']])
```

```python
train.isna().sum()
```

```python
train = train.drop('id', axis = 1)
train.isna().sum()
```

```python
train.head()
```

```python
train['DC201'] = train.DC201.map({'Layak Minum': 1, 'Tidak Layak Minum': 0})
```

```python
feat_data = train.drop(columns =['DC201'])
feat_data.head()
```

```python
label_data = train['DC201']
label_data.head()
```

#Split data

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(feat_data, label_data, test_size=0.2, random_state=500)
```

#Standarization

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(xtrain)
Xtrain_std = sc.transform(xtrain)
Xtest_std = sc.transform(xtest)
```

#DECISION TREE

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.tree import DecisionTreeClassifier
from sklearn import datasets
from sklearn import metrics

dt_cl = DecisionTreeClassifier (random_state =500)
#fit the model
dt_cl.fit (Xtrain_std, ytrain)

# Make the predictions
y_preddt_test = dt_cl.predict(Xtest_std)
y_preddt_train= dt_cl.predict(Xtrain_std)
 
# Measure the performance
print("Accuracytest score %.3f" %metrics.accuracy_score(ytest, y_preddt_test))
print("Accuracytrain score %.3f" %metrics.accuracy_score(ytrain, y_preddt_train))
```

#SVM

```python
# Sklearn modules & classes
from sklearn.linear_model import Perceptron, LogisticRegression
from sklearn.svm import SVC

# Instantiate the Support Vector Classifier (SVC)
svc = SVC(C=1.0, random_state=1, kernel='linear')
 
# Fit the model
svc.fit(Xtrain_std, ytrain)

# Make the predictions
y_predict = svc.predict(Xtest_std)
y_predicttrain= svc.predict(Xtrain_std)
 
# Measure the performance
print("Accuracytest score %.3f" %metrics.accuracy_score(ytest, y_predict))
print("Accuracytrain score %.3f" %metrics.accuracy_score(ytrain, y_predicttrain))
```

#NAIVE BAYES

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()

# Fit the model
nb.fit (Xtrain_std, ytrain)

# Make the predictions
y_prednb= nb.predict(Xtest_std)
y_predtrainnb= nb.predict(Xtrain_std)
 
# Measure the performance
print("Accuracytest score %.3f" %metrics.accuracy_score(ytest, y_prednb))
print("Accuracytrain score %.3f" %metrics.accuracy_score(ytrain, y_predtrainnb))
```

#Logistic regression

```python
from sklearn.linear_model import LogisticRegression

# Fit the model
logreg = LogisticRegression(solver='liblinear', random_state=0)
logreg.fit(feat_data, label_data)

# Make the predictions
y_predlr= logreg.predict(Xtest_std)
y_predtrainlr= logreg.predict(Xtrain_std)

# Measure the performance
print("Accuracytest score %.3f" %metrics.accuracy_score(ytest, y_predlr))
print("Accuracytrain score %.3f" %metrics.accuracy_score(ytrain, y_predtrainlr))
```

#Random Forest

```python
# Instantiate the Support Vector Classifier (SVC)
from sklearn import ensemble
rf_cl = ensemble.RandomForestClassifier (n_estimators=100)

# Fit the model
rf_cl.fit (Xtrain_std, ytrain)

# Make the predictions
y_predrfcl= rf_cl.predict(Xtest_std)
y_predtrainrfcl= rf_cl.predict(Xtrain_std)
 
# Measure the performance
print("Accuracytest score %.3f" %metrics.accuracy_score(ytest, y_predrfcl))
print("Accuracytrain score %.3f" %metrics.accuracy_score(ytrain, y_predtrainrfcl))
```

#Gradient Boosting

```python
from sklearn.metrics import classification_report
from sklearn.datasets import load_breast_cancer
from sklearn.ensemble import GradientBoostingClassifier

from sklearn.metrics.pairwise import gen_batches
gb= GradientBoostingClassifier(learning_rate=0.1)
#fit the model
gb.fit (Xtrain_std, ytrain)

# Make the predictions
y_predgb= gb.predict(Xtest_std)
y_predtraingb= gb.predict(Xtrain_std)
 
# Measure the performance
print("Accuracytest score %.3f" %metrics.accuracy_score(ytest, y_predgb))
print("Accuracytrain score %.3f" %metrics.accuracy_score(ytrain, y_predtraingb))
```

#KNN

```python
from sklearn.neighbors import KNeighborsClassifier

knn_cl = KNeighborsClassifier (n_neighbors=3)
#fit the model
knn_cl.fit (Xtrain_std, ytrain)
# Make the predictions
y_predknn_cl= knn_cl.predict(Xtest_std)
y_predtrainknn_cl= knn_cl.predict(Xtrain_std)
 
# Measure the performance
print("Accuracytest score %.3f" %metrics.accuracy_score(ytest, y_predknn_cl))
print("Accuracytrain score %.3f" %metrics.accuracy_score(ytrain, y_predtrainknn_cl))
```
